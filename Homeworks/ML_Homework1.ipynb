#### *How would you define Machine Learning?*
> Machine Learning is the process of computers to learn without directly being typed, instead the programmer gives input and output data into the required function and then creates models. These models has the ability to predict.

#### *What are the differences between Supervised and Unsupervised Learning? Specify example 3 algorithms for each of these.*
> In the case of **Supervised Learning** we give the input and output data, then we expect the algorithm. On the other hand, in **Unsupervised Learning** we only provide the input data, and then computer tries to make predictions.

> *Supervised:*
1. Linear Regression
2. Logical Regression
3. Ensemble Methods

> *Unsupervised:*
1. Clustering
2. Association
3. K Means

#### *What are the test and validation set, and why would you want to use them?*
>While we are trying to create a machine learning model we need to split our data.

> This can done in two ways:
- 2 parts => Train{70%} + Test{30%}
- 3 parts => Train{60%} + Validation{20%} + Test{20%} 

> **Test** set is used to check our model accuracy, wheter it underfits or overfits?

> **Validation** set is used to improve the model created by the Train set.
